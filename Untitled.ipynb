```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
metrics = ['Accuracy', 'Precision (0)', 'Precision (1)', 'Recall (0)', 'Recall (1)', 'F1-Score (0)', 'F1-Score (1)']
linear_regression = [0.78, 0.79, 0.76, 0.85, 0.69, 0.82, 0.72]
decision_trees = [0.7989, 0.83, 0.75, 0.82, 0.77, 0.83, 0.76]
```

```python
data = pd.DataFrame({
    'Metrics': metrics,
    'Linear Regression': linear_regression,
    'Decision trees': decision_trees
})
```

```python
data_melted = data.melt(id_vars='Metrics', var_name='Model', value_name='Score')
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Metrics', y='Score', hue='Model', data=data_melted)
```
